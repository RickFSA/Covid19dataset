### 3. Prepare the data for clustering

1. Remove the `sex`, `sport` column. We will see if the clusters from K-Means end up like the actual sex.
2. Put the features on the same scale.

```python
# ANSWER
df = ais.drop(['sex', 'sport'], axis=1)
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns) 
```

```python
scaled_df.head(1)
```

### 4. Clustering with K-Means

- Cluster the data to our our target groups.
    - Use sklearn.cluster.KMeans
- We know that there are 2 actual classes. Lets try using 2, 4. Run the clustering algorithm with each of those k values.

```python
from sklearn.cluster import KMeans
```

```python
kmek = KMeans(n_clusters=2)
kmek.fit(scaled_df)
```

```python
# Class definition
#class Cluster:
    def __init__(self):
        pass
    
    def kmeans(self, n_clusters, data):
        kmean = KMeans(n_clusters=n_clusters)
        kmean.fit(data)
        labels = kmean.labels_
        return labels, kmean
    
    def elbow(self, data):
        distance = []
        k_range = range(1,15)
        for k in k_range:
            km = KMeans(n_clusters=k)
            km.fit(data)
            distance.append(km.inertia_)
        return k_range, distance
    
    def plot_elbow(self, k_range, distance):
        sns.lineplot(k_range, distance, markers='>')
        plt.xlabel('Number of Cluster'); plt.ylabel('Distance')
        plt.title('Elbow Method for Optimal K')
        plt.show()
                
```

### 5. Get the labels for the models

```python
# ANSWER
labels = kmek.labels_
labels
```

### 6. Elbow Method For Optimal k

For each k value, we will initialise k-means and use the inertia attribute to identify the sum of squared distances of samples to the nearest cluster centre. As k increases, the sum of squared distance tends to zero. 

```python
# ANSWER
distance = []
rangek = range(1,10)
for k in rangek:
    km = KMeans(n_clusters=k)
    km.fit(scaled_df)
    distance.append(km.inertia_)
```

```python
distance
```

Below is a plot of sum of squared distances for k in the range specified above. If the plot looks like an arm, then the elbow on the arm is optimal k.

```python
# ANSWER
sns.set_style('darkgrid')
plt.figure(figsize=(8,6))
plt.plot(rangek, distance, marker='o')
```
