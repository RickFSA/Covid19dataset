```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context('poster')
```

The data can be find on https://en.climate-data.org/oceania/australia/new-south-wales/sydney-24/

```python
rain = [['Jan', 126], ['Feb', 147], ['Mar', 155], ['Apr', 112], ['May', 110], ['Jun', 152], ['Jul', 74], ['Aug',93], ['Sep', 60], ['Oct', 92], ['Nov', 103], ['Dec', 85]]
df = pd.DataFrame(rain, columns =['month', 'rainfall'])
df
```

```python
import seaborn as sns
sns.set_style('whitegrid')
plt.figure(figsize=(15,10))
plt.bar(df.month, df.rainfall)
plt.xlabel('Time')
plt.ylabel('Rainfall in (mm)')
plt.title('Sydney Average Rainfall (1982-2012)')
plt.savefig('rainfall.png')
```
