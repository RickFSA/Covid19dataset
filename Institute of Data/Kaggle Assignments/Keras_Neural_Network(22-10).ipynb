```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import EarlyStopping
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv', delimiter=',', header=None)
```

```python
df.columns = ['pregnant', 'glucose', 'blood_pressure', 'skin_thickness', 'insulin', 'bmi', 'diabetes_function', 'age', 'class']
df.head()
```

## Preprocessing

```python
X = df.iloc[:,:8]
y = df.iloc[:,-1]
```

```python
# Scale data and transform variables to array matrix
scaler = StandardScaler()
Xs = scaler.fit_transform(X)
Xs
```

```python
y = np.array(y)
y[:10]
```

## Neural Network Model

```python
model = Sequential()

model.add(Dense(50, activation='relu', input_shape=(Xs.shape[1],)))
model.add(Dense(20, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(Xs, y, epochs=200, batch_size=20, validation_split=0.3, class_weight={0:.3,1:.7}, verbose=1, )
```

```python
predictions = model.predict(Xs).round()
```

```python
# Alternatively, we can use model.predict_classes(x)
label = model.predict_classes(Xs)
```

```python
model.summary()
```

```python
print(f'Model accuracy score: {accuracy_score(y, predictions):.2f}')
```
